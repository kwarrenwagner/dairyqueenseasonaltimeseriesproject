## Identifying potential series components for model evaluation:

```r
install.packages("tseries")
library(tseries)
```

```r
dairy_queendf_ts_train <- ts(train$dq_search_interest,
                       start = c(year(min(train$date)), month(min(train$date))),
                       frequency = 12)
dairy_queendf_ts_test <- ts(test$dq_search_interest,
start = c(year(min(test$date)), month(min(test$date))),
frequency = 12)
```

```r
decomposition <- decompose(dairy_queendf_ts_train)
plot(decomposition)
```

```r
differenceddq<-diff(diff(dairy_queendf_ts_train,lag=12))
```

```r
acf(differenceddq, main = "ACF of Dairy Queen Search Interest", lag.max = 48)
pacf(differenceddq, main = "PACF of Dairy Queen Search Interest", lag.max = 48)
```

Spike at lag one in PACF maybe AR(1) component, not really anything seasonal.There is a potential seasonal trend in ACF plot although it could be argued it is just decaying and a spike at lag one, so maybe MA(1) component for seasonal and non-seasonal. We will use this insights to choose potential models. Given these seasonal patterns, it seems fair to use SARIMA models, Holt Winters forecasting to predict future values.

```r
install.packages("forecast")
library(forecast)
```

## SARIMA (choosing candidates based on ACF and PACF):

These are the candidate models

1. SARIMA(1, 1, 1)(0, 0, 0)[12]:

AR(1) and MA(1) for the non-seasonal components, no seasonal components. This is a simple model capturing the identified AR and MA at lag 1.

2. SARIMA(1, 1, 1)(0, 1, 0)[12]:

AR(1) and MA(1) for non-seasonal, seasonal differencing (D=1) to address the potential seasonal trend in the ACF.

3. SARIMA(1, 1, 1)(0, 0, 1)[12]:

AR(1) and MA(1) for non-seasonal, seasonal MA(1) (Q=1) to capture potential seasonal autocorrelation in the ACF.

4. SARIMA(1, 1, 1)(0, 1, 1)[12]:

AR(1) and MA(1) for non-seasonal, seasonal differencing (D=1), and seasonal MA(1) (Q=1). This combines both approaches to handle the potential seasonal trend.

5. SARIMA(1, 1, 0)(0, 0, 1)[12]:

AR(1) for non-seasonal, non-seasonal MA is excluded, seasonal MA(1) (Q=1). This tests if the MA at lag 1 is truly needed alongside a seasonal MA.

6. SARIMA(0, 1, 1)(0, 0, 1)[12]:

Non-seasonal MA(1) only, seasonal MA(1) (Q=1). This tests if the AR at lag 1 is truly needed alongside a seasonal MA.

```r
#1. SARIMA(1,1,1)(0,0,0)[12]
fit1 <- Arima(dairy_queendf_ts_train, order = c(1,1,1), seasonal = list(order = c(0,0,0), period = 12))

#2. SARIMA(1,1,1)(0,1,0)[12]
fit2 <- Arima(dairy_queendf_ts_train, order = c(1,1,1), seasonal = list(order = c(0,1,0), period = 12))

#3. SARIMA(1,1,1)(0,0,1)[12]
fit3 <- Arima(dairy_queendf_ts_train, order = c(1,1,1), seasonal = list(order = c(0,0,1), period = 12))

#4. SARIMA(1,1,1)(0,1,1)[12]
fit4 <- Arima(dairy_queendf_ts_train, order = c(1,1,1), seasonal = list(order = c(0,1,1), period = 12))

#5. SARIMA(1,1,0)(0,0,1)[12]
fit5 <- Arima(dairy_queendf_ts_train, order = c(1,1,0), seasonal = list(order = c(0,0,1), period = 12))

#6. SARIMA(0,1,1)(0,0,1)[12]
fit6 <- Arima(dairy_queendf_ts_train, order = c(0,1,1), seasonal = list(order = c(0,0,1), period = 12))
```

```r
#AICc in r wasn't working
calculate_AICc <- function(fit, n_obs) {
  k <- length(fit$coef) + 1
  AIC_val <- AIC(fit)
  AICc_val <- AIC_val + (2 * k * (k + 1)) / (n_obs - k - 1)
  return(AICc_val)
}
```

```r
n <- length(dairy_queendf_ts_train)

calculate_AICc(fit1, n)
calculate_AICc(fit2, n)
calculate_AICc(fit3, n)
calculate_AICc(fit4, n)
calculate_AICc(fit5, n)
calculate_AICc(fit6, n)
```

4. SARIMA(1,1,1)(0,1,1)[12] appears to be the best model based on AICC

```r
model1forecast <-forecast(fit4)
```

```r
plot(model1forecast)
```

```r
rmse <- accuracy(model1forecast, dairy_queendf_ts_test)["Test set", "RMSE"]
mape <- accuracy(model1forecast, dairy_queendf_ts_test)["Test set", "MAPE"]
```

```r
rmse
mape
```
